```python
# Agregamos columna velocidad por trayecto 
df_vuelo = df_vuelos.assign(Mediavelocidad = df_vuelos['Distance'] / df_vuelos['DepTime']) 
df_vuelo
```

```python
# Agregamos columna que nos indica si llega tarde
df_vuelo = df_vuelo.assign(Llegatarde = df_vuelo['ArrDelay'] + df_vuelo['ArrTime'] > df_vuelo['ArrTime'])
df_vuelo
```

```python
# Creo tabla de retrasos con la aerolinia
df_retrasos = df_vuelo.loc[:,['UniqueCarrier', 'ArrDelay','DepDelay','CarrierDelay',
       'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay','Origin', 'Dest', 'Distance','Mediavelocidad','Llegatarde']]
df_retrasos
```

```python
# Añado columna = suma total retraso por cada vuelo.
df_retrasos['Totalretrasos'] = df_retrasos['WeatherDelay'] + df_retrasos['ArrDelay'] + df_retrasos['DepDelay'] + df_retrasos['CarrierDelay'] + df_retrasos['NASDelay'] + df_retrasos['SecurityDelay'] + df_retrasos['LateAircraftDelay']

df_retrasos    
```

```python
# Sumo columna totales y agrupo por cada aerolinia y selecciono las 10 que más retrasos tienen.
df_retraso_aerolinia = df_retrasos.groupby('UniqueCarrier')['Totalretrasos'].sum().reset_index()

resultados_max = df_retraso_aerolinia.nlargest(10, 'Totalretrasos')

print(resultados_max)
```

```python
# Ordeno la columna distancia en orden descendiente 
rutas = df_vuelo.sort_values(by=['Distance'], ascending=False)
```

```python
# mostrar las 60 rutas más largas

rutas_mas_largas = rutas.head(60)
print(rutas_mas_largas)
```

```python
# Añadí una nueva columna de rutas porque entiendo que en este listado se repiten los vuelos, intenté agrupar con groupby por rutas, pero no me ha salido, después me decia que podía utilizar drop para eliminar duplicados pero no quería que afectara a la base de datos.
# Así que lo entrego de este modo. Pero lo vuelvo a intentar...
```

```python
# Agregar columna nueva ruta
df_retrasos['Ruta'] = df_retrasos['Origin'] + ' - ' + df_retrasos['Dest']
df_retrasos
```

```python
# Agrupo por cada aerolinia y selecciono las 20 rutas más largas.
df_distancia_ruta = df_retrasos.groupby('Ruta') ['Distance'].first().reset_index()

resultados_rutas_largas = df_distancia_ruta.nlargest(20, 'Distance')

print(resultados_rutas_largas)
```

```python
# Muestro las 20 rutas que acumulan más retrasos.
df_distancia_ruta = df_retrasos.groupby('Ruta') ['Totalretrasos'].first().reset_index()

resultados_rutas_retraso = df_distancia_ruta.nlargest(20, 'Totalretrasos')

print(resultados_rutas_retraso)
```
